## Depth First Search

* Directed Graph

```python
edges=[('a','b'),('a','c'),('a','e'),('a','d'),
('b','e'),
('c','g'),('c','d'),('c','e'),
('d','f'),
('e','f'),
('f','g')]
```

```python
g=create_graph(edges,directed=False)
```

```python
g.getNode('a').connectedTo
```

```python
def DFS(g,top_node):
    q = deque()
    top_node=g.nodes[top_node]
    top_node.state=1
    q.append(top_node)
    top_node.distance=0
    while len(q)>0:
        node=q.popleft()
        if node.state!=-1:
            for connected_node in node.connectedTo:
                connected_node=g.nodes[connected_node]
                if connected_node.state==0:
                    connected_node.state=1
                    connected_node.distance=node.distance+1
                    q.append(connected_node)
            node.state=-1
        print('distnce from the node',node.key,node.distance)    
```

```python
g=create_graph(edges,directed=False)
DFS(g,'a')
```

```python
g=create_graph(edges,directed=False)
DFS(g,'f')
```

```python
g=create_graph(edges,directed=False)
DFS(g,'d')
```
